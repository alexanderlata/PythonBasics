## Локальный минимум функции одной переменной

```python
from scipy.optimize import minimize_scalar

# Определение функции
def func(x):
    return x**2 - 4*x + 4  # Квадратичная функция

# Нахождение минимума функции с использованием метода золотого сечения
result = minimize_scalar(func, method='golden')

print("Минимум функции:", result.fun)
print("Точка минимума:", result.x)
```

### Поиск локального минимума на интервале 

Метод `Bounded` может выполнять ограниченную минимизацию. 

Он использует метод Брента для поиска локального минимума в интервале $x_1 < x_{opt} < x_2$, где $x_{opt} -$ оптимальное решение.

```python
from scipy.optimize import minimize_scalar

# Определение функции
def func(x):
    return x**2 - 4*x + 4  # Квадратичная функция

# Нахождение минимума функции с использованием метода 
result = minimize_scalar(func, method='Bounded', bounds=(-5, 1))

print("Минимум функции:", result.fun)
print("Точка минимума:", result.x)
```

```python
import numpy as np
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as plt

# Определение функции с тремя точками минимума
def func(x):
    return np.sin(x) + 0.5*np.sin(2*x) + 0.1*np.sin(3*x)

# Нахождение минимума функции с использованием метода Ньютона
result = minimize_scalar(func, method='bounded', bounds=(-5, 5))

print("Минимум функции:", result.fun)
print("Точка минимума:", result.x)

# Построение графика функции
x_values = np.linspace(-5, 5, 100)
y_values = func(x_values)

plt.plot(x_values, y_values, label='Function')
plt.scatter(result.x, result.fun, color='red', label='Minimum')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Function with infinity local minima')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize

# Определение функции
def func(x):
    return (x ** 3) / (2 * ((x + 1) ** 2))

# Нахождение минимума функции с использованием метода Nelder-Mead
res = minimize(func, x0=1, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})
res
```

```python
print("Минимум функции:", res.fun)
print("Точка минимума:", res.x)
```

## Локальный максимум функции одной переменной


Для поиска максимума функции одной переменной можно использовать те же методы, что и для поиска минимума, но просто умножить функцию на -1. Это превратит задачу поиска максимума в задачу поиска минимума.

```python
from scipy.optimize import minimize_scalar
import numpy as np

# Определение функции
def func(x):
    return -np.sin(x) * x**2  # Умножаем на -1 для поиска максимума

# Нахождение максимума функции с использованием метода Ньютона
result = minimize_scalar(func, method='bounded', bounds=(-5, 5))

print("Максимум функции:", -result.fun)  # Возвращаем значение функции к положительному
print("Точка максимума:", result.x)
```

## Глобальный минимум функции одной переменной

```python
from scipy.optimize import differential_evolution

# Определение функции
def func(x):
    return x**4 - 3*x**3 + 2*x**2 - x + 1

# Нахождение глобального минимума функции с использованием метода дифференциальной эволюции
result = differential_evolution(func, bounds=[(-5, 5)])

print("Глобальный минимум функции:", result.fun)
print("Точка глобального минимума:", result.x)
```
